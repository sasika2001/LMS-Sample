```python
from pymongo import MongoClient

# Replace this with your actual MongoDB URI
mongo_uri = "mongodb+srv://admin:1234@cluster0.ogyx0.mongodb.net/LMS?retryWrites=true&w=majority&appName=Cluster0"

# Connect to MongoDB
client = MongoClient(mongo_uri)

# Select your database
db = client["LMS"]

# Test connection: list collections
db.list_collection_names()
```

```python
import pandas as pd

def load_logins(collection_name):
    cursor = db[collection_name].find({}, {
        "username": 1,
        "role": 1,
        "timestamp": 1,
        "location": 1,
        "device_info": 1,
        "hour": 1
    })
    df = pd.DataFrame(list(cursor))
    df["source"] = collection_name
    return df

# Load from all collections
student_df = load_logins("student_login_details")
teacher_df = load_logins("teacher_login_details")
admin_df = load_logins("admin_login_details")

# Combine them
all_logins_df = pd.concat([student_df, teacher_df, admin_df], ignore_index=True)

# Drop rows with missing values in key columns
all_logins_df.dropna(subset=["username", "role", "timestamp", "location", "device_info"], inplace=True)

# Preview the cleaned DataFrame
all_logins_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Helper to extract hour from timestamp
def extract_hour(timestamp):
    if isinstance(timestamp, str):
        try:
            return pd.to_datetime(timestamp).hour
        except Exception:
            return None
    return timestamp.hour if hasattr(timestamp, 'hour') else None

# Preprocess the DataFrame
all_logins_df["hour"] = all_logins_df["timestamp"].apply(extract_hour)
all_logins_df["location"] = all_logins_df["location"].apply(lambda x: x.split(",")[0].strip() if isinstance(x, str) else "Unknown")
all_logins_df["device_info"] = all_logins_df["device_info"].apply(lambda x: x[:50] if isinstance(x, str) else "Unknown")

# Drop rows where hour couldn't be extracted
all_logins_df.dropna(subset=["hour"], inplace=True)

# Encode categorical features
location_encoder = LabelEncoder()
device_encoder = LabelEncoder()

all_logins_df["location_encoded"] = location_encoder.fit_transform(all_logins_df["location"])
all_logins_df["device_encoded"] = device_encoder.fit_transform(all_logins_df["device_info"])

# Final dataset for each model
X_time = all_logins_df[["hour"]]
X_location = all_logins_df[["location_encoded"]]
X_device = all_logins_df[["device_encoded"]]

# Preview processed data
all_logins_df[["username", "hour", "location", "location_encoded", "device_info", "device_encoded"]].head()
```

```python
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style for plots
sns.set(style="whitegrid")

# === Train models ===
time_model = IsolationForest(contamination=0.1, random_state=42)
location_model = IsolationForest(contamination=0.1, random_state=42)
device_model = IsolationForest(contamination=0.1, random_state=42)

time_model.fit(X_time)
location_model.fit(X_location)
device_model.fit(X_device)

# === Predict anomalies ===
all_logins_df["time_anomaly"] = time_model.predict(X_time)
all_logins_df["location_anomaly"] = location_model.predict(X_location)
all_logins_df["device_anomaly"] = device_model.predict(X_device)

# Convert -1 (anomaly) / 1 (normal) to boolean
all_logins_df["is_time_anomaly"] = all_logins_df["time_anomaly"] == -1
all_logins_df["is_location_anomaly"] = all_logins_df["location_anomaly"] == -1
all_logins_df["is_device_anomaly"] = all_logins_df["device_anomaly"] == -1

# Overall anomaly if any one of the models flags it
all_logins_df["is_overall_anomaly"] = (
    all_logins_df["is_time_anomaly"] |
    all_logins_df["is_location_anomaly"] |
    all_logins_df["is_device_anomaly"]
)

# === Summary ===
print("Total records:", len(all_logins_df))
print("Total overall anomalies detected:", all_logins_df["is_overall_anomaly"].sum())
```

```python
# Pie chart of anomalies vs. normal
labels = ['Anomaly', 'Normal']
colors = ['#FF6F61', '#6BA292']

plt.figure(figsize=(14, 4))

# Time anomaly
plt.subplot(1, 3, 1)
time_counts = all_logins_df["is_time_anomaly"].value_counts()
plt.pie(time_counts, labels=labels, autopct='%1.1f%%', colors=colors)
plt.title("Time Anomaly")

# Location anomaly
plt.subplot(1, 3, 2)
loc_counts = all_logins_df["is_location_anomaly"].value_counts()
plt.pie(loc_counts, labels=labels, autopct='%1.1f%%', colors=colors)
plt.title("Location Anomaly")

# Device anomaly
plt.subplot(1, 3, 3)
dev_counts = all_logins_df["is_device_anomaly"].value_counts()
plt.pie(dev_counts, labels=labels, autopct='%1.1f%%', colors=colors)
plt.title("Device Anomaly")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(
    all_logins_df[["is_time_anomaly", "is_location_anomaly", "is_device_anomaly"]].corr(),
    annot=True, cmap="coolwarm", fmt=".2f"
)
plt.title("Correlation Between Anomaly Types")
plt.show()
```

```python
import time
import numpy as np
import matplotlib.pyplot as plt
from pymongo import MongoClient
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import LabelEncoder
from IPython.display import clear_output

# MongoDB connection
client = MongoClient("mongodb+srv://admin:1234@cluster0.ogyx0.mongodb.net/LMS?retryWrites=true&w=majority&appName=Cluster0")
db = client["LMS"]

# Fetch login data from all roles
def fetch_login_data():
    collections = ["student_login_details", "teacher_login_details", "admin_login_details"]
    devices, locations, hours = [], [], []
    for col in collections:
        for doc in db[col].find({}, {"device_info": 1, "location": 1, "hour": 1}):
            if doc.get("device_info"):
                devices.append(doc["device_info"][:50])
            if doc.get("location"):
                locations.append(doc["location"].split(",")[0].strip())
            if doc.get("hour") is not None:
                hours.append(doc["hour"])
    return devices, locations, hours

# Train models
def train_model(data):
    encoder = LabelEncoder()
    encoded = encoder.fit_transform(data).reshape(-1, 1)
    model = IsolationForest(contamination=0.1, random_state=42)
    model.fit(encoded)
    return model, encoder

# Predict anomalies
def predict_anomalies(model, encoder, data):
    encoded = encoder.transform(data).reshape(-1, 1)
    scores = model.decision_function(encoded)
    preds = model.predict(encoded)
    return preds, scores

# Real-time dashboard loop
while True:
    clear_output(wait=True)
    print("Fetching new login data and updating graphs...")

    # Step 1: Fetch data
    devices, locations, hours = fetch_login_data()
    if not devices or not locations or not hours:
        print("Waiting for enough login data...")
        time.sleep(30)
        continue

    # Step 2: Train models
    device_model, device_enc = train_model(devices)
    location_model, loc_enc = train_model(locations)
    time_model = IsolationForest(contamination=0.1, random_state=42).fit(np.array(hours).reshape(-1, 1))

    # Step 3: Predict
    dev_preds, dev_scores = predict_anomalies(device_model, device_enc, devices)
    loc_preds, loc_scores = predict_anomalies(location_model, loc_enc, locations)
    time_preds = time_model.predict(np.array(hours).reshape(-1, 1))
    time_scores = time_model.decision_function(np.array(hours).reshape(-1, 1))

    # Step 4: Visualization
    fig, axs = plt.subplots(1, 2, figsize=(14, 5))

    # Anomaly count bar chart
    counts = {
        'Device': list(dev_preds).count(-1),
        'Location': list(loc_preds).count(-1),
        'Time': list(time_preds).count(-1),
    }
    axs[0].bar(counts.keys(), counts.values(), color=["tomato", "orange", "skyblue"])
    axs[0].set_title("Anomaly Count by Feature")
    axs[0].set_ylabel("Count")

    # Anomaly score histogram
    all_scores = np.concatenate([dev_scores, loc_scores, time_scores])
    axs[1].hist(all_scores, bins=50, color='purple', alpha=0.7)
    axs[1].set_title("Anomaly Score Distribution")
    axs[1].set_xlabel("Score")
    axs[1].set_ylabel("Frequency")

    plt.tight_layout()
    plt.show()

    print("Waiting for next update...")
    time.sleep(30)  # refresh every 30 seconds
```
